```python
import pandas as pd
import matplotlib.pyplot as plt

# Define your data
data = {
    'Storage Size': ['No storage', 6, 9, 14, 28, 42],
    'OPEX': [539.66, 573.75, 801.76, 988.36, 1127.74, 1411.53][::-1],
    'CAPEX': [835.5, 626, 416.5, 390, 361.25, 207][::-1],
    'TOTEX': [1375.16, 1199.75, 1218.26, 1378.36, 1488.99, 1618.53][::-1]
}
df = pd.DataFrame(data)

# Extract data for plotting
storage_sizes = df['Storage Size']
opex = df['OPEX']
capex = df['CAPEX']
totex = df['TOTEX']

# Define the positions for the bars
x = range(len(storage_sizes))

# Create a stacked bar plot
plt.figure(figsize=(6, 4))

# Plotting the stacked bars
plt.bar(x, capex, width=0.4, label='CAPEX', color='tab:blue')
plt.bar(x, opex, bottom=capex, width=0.4, label='OPEX', color='tab:orange')
plt.plot(totex, color='black',label='TOTEX')
plt.scatter(x,totex, color='black')

# Add labels, title, and ticks
plt.xlabel('Storage capacity [kWh]')
plt.ylabel('Yearly costs [$/year]')
plt.xticks(x, storage_sizes)
plt.legend()

# Show plot
plt.tight_layout()
plt.savefig('plot.png', dpi=300)
plt.show()
```
